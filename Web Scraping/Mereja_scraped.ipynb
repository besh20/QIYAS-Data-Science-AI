```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://mereja.net/amharic/'

res = requests.get(url)
data = BeautifulSoup(res.text, 'html.parser')
data
```

```python
records = []

for post in data.find_all("article"):

    meta = post.find("div", class_="post-meta")
    content = post.find("div", class_="entry-content")
    post_entry = post.find("div", class_="post-entry")

    records.append({
        "title": post.find("h2").get_text(strip=True),
        "author": meta.find("span", class_="author").get_text(strip=True) ,
        "date": meta.find("time").get_text(strip=True),
        "content": content.get_text(" ", strip=True),
        "url": post_entry.find("a")["href"]
    })

df = pd.DataFrame(records)
df
```

```python
df.to_csv('scraped_mereja.txt', index=False)
```
