```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies= pd.read_csv('movies_updated.csv')
movies.head(3)
```

### Task Zero: Cleaning 

```python
movies['runtime,,'] = movies['runtime,,'].str.replace(r',', '' , regex=False)
movies = movies.rename(columns={'runtime,,': 'runtime'})
```

```python
movies['runtime'] = movies['runtime'].astype(float)
```

### Task One: Exploration 

```python
print(f'This movie dataset contains {len(movies)} movies')
print(f'Numerical columns: {movies.select_dtypes(include=['number']).columns}.The others are categorical')
```

### Task Two: Number of Movies Released Each Year

```python
yearly_counts = movies['year'].value_counts()
plt.figure(figsize=(8, 3))
yearly_counts.plot(kind='line');
```

Years from 1985 and above have the same number movies released(200). Rather than increasing it stays constant from that specific year on. 

### Task Three: Top 10 Movie Genres

```python
top_genre = movies['genre'].value_counts().head(10)
top_genre.plot(kind='bar');
```

mystery: lowest, Comedy highest. the bar visual is suitable for comparing categories.

### Task Four: Top 10 Directors

```python
top_dir = movies['director'].value_counts().head(10)
top_dir.plot(kind='barh');
```

Woody ALlen leads the way to direct the largest number of movie. since the length of the directors name are long barh is more suitable

### Task Five: Distribution of Movie Ratings

```python
score_dis = movies['score']
score_dis.plot(kind='hist');
```

Score 6-7(6.5-7) has the highest number movies. its okay to say most of the movies are neither high nor low rated rather fairly rated.

### Task Six: Average Movie Score by Genre

```python
genre_score = movies.groupby('genre')['score'].mean().sort_values()
genre_score.plot(kind='barh');
```

highest: History, lowest: Western. Its interesting to not see the popular genere dominate by score. 

### Task Seven: Top 10 Highest Grossing Movies

```python
top_gross = movies.sort_values(by='gross', ascending=False).head(10)
top_gross.plot(kind='barh',x='name', y='gross');
```

Titanic wipes all the others with a whopping 2.8$ billion grossing worldwide which is a lot from the 2nd on the list.
