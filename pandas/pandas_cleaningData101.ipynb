```python
import pandas as pd
import numpy as np

uni = pd.DataFrame({
    "student_id": [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    "name": ["Abel", "Sara", None, "John", "Marta", None, "David", "Helen", "Tom", None],
    "department": ["CS", "IT", "CS", None, "SE", "IT", "SE", None, "CS", "IT"],
    "gpa": [3.5, 3.8, np.nan, 2.9, 3.7, np.nan, 3.2, 3.4, 3.9, np.nan],
    "scholarship": [5000, np.nan, 3000, 2000, np.nan, 4000, 3500, np.nan, 5000, np.nan]
})
uni
```

#### DATA CLEANING 


```python
# number of missing values in this dataframe 
uni.isna().sum().sum() 
```

```python
#  percentages of missing values 
(uni.isna().sum() / len(uni) ) * 100 
```

```python
# replace missing values with unknown in the name column.   
uni['name'] = uni['name'].fillna('Unknown')
```

```python
# relpace with the mode of department. use index if there are multiple modes.
uni['department'] = uni['department'].fillna(uni['department'].mode()[0])
```

```python
# replace with the median of gpa. 
uni['gpa'] = uni['gpa'].fillna(uni['gpa'].median())
```

```python
# replace missing values with 0 in the scholarship column
uni['scholarship'] = uni['scholarship'].fillna(0) 
```

### DATA ANALYSIS 

```python
# the number of students by departments
uni['department'].value_counts()
```

```python
# students who received scholarships 
(uni['scholarship'] != 0).sum()
```
