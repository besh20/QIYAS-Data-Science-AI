```python
df8 = pd.DataFrame({
    "salary": [3500,4000,4200,3900,4100,4500,93424]
})
Q1 = df8['salary'].quantile(.25)
Q3 = df8['salary'].quantile(.75)

IQR = Q3 - Q1
lower  = Q1 - 1.5 * IQR
upper  = Q3 + 1.5 * IQR

outliers = df8[
    (df8['salary'] < lower) | (df8['salary'] > upper)
    ]

outliers

clean_df = df8[
    (df8['salary'] >= lower) & (df8['salary'] <= upper)
    
]
clean_df
```
