### Section D: Data Cleaning

```python
df2.isna().sum()
df2.isna().sum().sum()
```

```python
df2['Age'] = df2['Age'].fillna(df2['Age'].median())
df2['Score'] = df2['Score'].fillna(df2['Score'].mean())
df2
```

### Section E: Integrated Data Wrangling 

```python
df4 = pd.DataFrame({
    "Name": ["Alex","Nati","Sami","Tedi","Sami"],
    "Age": [23,24,23,np.nan,23],
    "enterance": ["2026-01-05", "05/02/2026", "March 10, 2026", "2026.04.15", "March 10, 2026"]
})

df4.duplicated()
```

```python
df4.drop_duplicates(keep='last')
df4['Age'] = df4['Age'].fillna(df4['Age'].mean())
df4["enterance"] = pd.to_datetime(df4["enterance"], format='mixed')
df4
```

```python
df5 = pd.DataFrame({
    "sales": [3500,4000,4200,60000,3900,4100,4500,93424]
})
Q1 = df5['sales'].quantile(.25)
Q3 = df5['sales'].quantile(.75)

IQR = Q3 - Q1
lower  = Q1 - 1.5 * IQR
upper  = Q3 + 1.5 * IQR

outliers = df5[
    (df5['sales'] < lower) | (df5['sales'] > upper)
    ]

outliers
```
